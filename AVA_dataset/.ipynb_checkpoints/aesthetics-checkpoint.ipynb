```python
import tensorflow as tf

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import urllib
import cStringIO
from PIL import Image
from HTMLParser import HTMLParser

%matplotlib inline
plt.style.use('ggplot')
```

```python
animal_train_ids = pd.read_table("aesthetics_image_lists/animal_train.jpgl", header=None, delimiter=" ")[0].tolist()
```

```python
aesthetic_ratings = pd.read_table("AVA.txt", header=None, delimiter=" ")
```

```python
class CustomHTMLParser(HTMLParser):
    def __init__(self):
        HTMLParser.__init__(self)
        self.recording = 0

    def handle_starttag(self, tag, attrs):
        attrs_as_dict = dict(attrs)
        if (self.recording):
           self.recording += 1 
        if tag == "td" and "class" in attrs_as_dict and attrs_as_dict["class"] == "page-image":
            self.recording = 1
        if (self.recording == 3):
            self.data = attrs_as_dict["src"]
```

```python
scores = {}
rows, cols = aesthetic_ratings.shape
for i in xrange(rows):
    row = aesthetic_ratings.iloc[i].tolist()
    ratings = row[2:12]
    numbers = [ ratings[k] * (k + 1) for k in xrange(len(ratings)) ]
    scores[row[1]] = sum(numbers)
```

```python
max_score = 0;
min_score = 100000000;
max_id = 0;
min_id = 0;
images = {}
i = 0
for id in animal_train_ids[:]:
    print i
    # Step 1: Get the URL for an image on its html page.
    URL = "http://www.dpchallenge.com/image.php?IMAGE_ID=" + str(id)    
    html_page = urllib.urlopen(URL).read()
    parser = CustomHTMLParser()
    try:
        parser.feed(html_page) 
    except:
        print "ERROR";
        continue
    
    
    # Step 2: Open the image and convert it to an 800x800 img by adding a white background
    file = cStringIO.StringIO(urllib.urlopen(parser.data).read())
    img = Image.open(file)
    width, height = img.size
    background = Image.new('RGB', (800, 800), (255, 255, 255))
    bg_w, bg_h = background.size
    offset = ((bg_w - width) / 2, (bg_h - height) / 2)
    background.paste(img, offset)
    
    
    #img_array = np.asarray(background)
    #if (len(img_array.shape) != 3):
    #    continue # Why does this image not have 3 dimensions???

    #reshaped_image = img_array.reshape(800, 800, 3)
    images[id] = background
    
    if (id in scores and scores[id] > max_score):
        max_score = scores[id]
        max_id = id
    
    if (id in scores and scores[id] < min_score):
        min_score = scores[id]
        min_id = id
    
    i += 1
```

```python
print max_score, max_id, min_score, min_id
#plt.imshow(reshaped_image)
```

```python
reshaped_images = {}
for image_id in images:
    img_array = np.asarray(images[image_id])
    if (len(img_array.shape) != 3):
        continue # Why does this image not have 3 dimensions???

    reshaped_image = img_array.reshape(800, 800, 3)
    reshaped_images[image_id] = reshaped_image
```

```python
normalized_scores = {}
for scores_id in scores:
    normalized_scores[scores_id] = float(scores[scores_id] - min_score) / float(max_score - min_score)
```

```python
# Build input tensors
x = tf.placeholder(tf.float32, shape=[None, 800 * 800 * 3])
y_ = tf.placeholder(tf.float32, shape=[None, 1])
```

```python
def conv_weight(shape):
    # conv weight shape [height, width, fan-in, fan-out]
    return tf.Variable(tf.truncated_normal(shape, stddev=1. / np.sqrt(shape[2])))
                       
def bias_weight(shape):
    return tf.Variable(tf.constant(0.1, shape=shape))
```

```python
# Transform vectors into an image
x_image = tf.reshape(x, [-1,800,800,3]) # [batch_count, height, width, channels]

# Set up variables for conv1
W_conv1 = conv_weight([5, 5, 3, 32]) # Filter [height, width, in channels, out channels]
b_conv1 = bias_weight([32])

# Set up convolution, relu and max pooling layer
conv1 = tf.nn.conv2d(x_image, W_conv1, strides=[1, 1, 1, 1], padding='SAME') + b_conv1
conv1 = tf.nn.relu(conv1)
pool1 = tf.nn.max_pool(conv1, ksize=[1, 2, 2, 1], strides=[1, 2, 2, 1], padding='SAME')
```

```python
# Print the shape for pool1 output
pool1.get_shape()
```

```python
def fc_weight(shape):
    # fc weight shape [fan-in, fan-out]
    return tf.Variable(tf.truncated_normal(shape, stddev=1. / np.sqrt(shape[0])))
    
# Set up variables for fc1
W_fc1 = fc_weight([400 * 400 * 32, 1])
b_fc1 = bias_weight([10])

# Flatten convolution response
pool1_flat = tf.reshape(pool1, [-1, 400 * 400 * 32])
fc1 = tf.matmul(pool1_flat, W_fc1) + b_fc1

# Get prediction probs from softmax 
pred = tf.nn.softmax(fc1)                 
```

```python
cross_entropy = - tf.reduce_sum(y_ * tf.log(pred))

# Training operation
train_op = tf.train.GradientDescentOptimizer(1e-4).minimize(cross_entropy)

# Evaluate prediction
correct_prediction = tf.equal(tf.argmax(pred, 1), tf.argmax(y_, 1))
accuracy = tf.reduce_mean(tf.cast(correct_prediction, "float"))
print correct_prediction
```

```python
with tf.Session() as sess:
    sess.run(tf.initialize_all_variables())
    batch = cnn_input
    print batch[0].shape
    train_accuracy = sess.run(accuracy, feed_dict={x:batch[0], y_: batch[1]}) # apply the "accuracy" tensor to inputs
    print "training accuracy %.2g," % (train_accuracy),
    #batch_val = mnist.validation.next_batch(50)
    #print "validation accuracy %.2g" % sess.run(accuracy, feed_dict={x: batch_val[0], y_: batch_val[1]})
    sess.run(train_op, feed_dict={x: batch[0], y_: batch[1]})

    #print "test accuracy %.2g" % sess.run(accuracy, feed_dict={x: mnist.test.images, y_: mnist.test.labels})
```
